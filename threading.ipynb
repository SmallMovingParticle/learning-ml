```python
import threading
import time
```

```python
def print_numbers():
    for i in range (5):
        print(f'the number is: {i}\n')

def print_letters():
    for letters in "abcdefight":
        print(f'the letter is: {letters} \n')

print_letters()

print_numbers()
```

```python
def print_numbers():
    for i in range (5):
        time.sleep(1)
        print(f'the number is: {i}\n')

def print_letters():
    for letters in "abcdefight":
        time.sleep(1)
        print(f'the letter is: {letters} \n')


t=time.time()
print_letters()

print_numbers()

finished_time = time.time()-t

print(finished_time)

##sleep is good as sleeping makes the other io operation time 

```

```python
##crating two threats

t1= threading.Thread(target=print_letters)

t2= threading.Thread(target=print_numbers)
t=time.time()

t1.start()
##when t1 will sleep t2 will be executing in parallel

t2.start()


t1.join()
t2.join()



finished_time = time.time()-t

print(finished_time)
```
