```python
## calculator

num1= int(input("enter the first number"))
num2= int(input("enter the second number"))

operator= input("enter thr operator + , - , * , / , %")

if operator=="+":
    print(num1+num2)
if operator=="*":
    print(num1*num2)
   
if operator=="/":
    print(num1/num2)

if operator=="-":
    print(num1-num2)
   
if operator=="%":
    print(num1%num2)
     
```

```python
n=int(input("enter the number"))
sum = 0
for i in range(1,n+1):
    sum = sum+i

print (sum)
```

```python
n=int(input("enter the number"))
sum = 0
count=0
while count<=n:
    sum= sum+count
    count+=1
print (sum)
```
