```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,6,8,4,5]

##create a line plot

plt.plot(x,y)

plt.xlabel('X axis')
plt.ylabel('Y lable')
plt.title('the first one')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,6,8,4,5]

##customized line plot

plt.plot(x,y,color='red', linestyle = '-.' , marker=
         'o', linewidth=4,markersize=7)


plt.grid(True)
```

```python
##multiplot
x=[1,2,3,4,5]
y=[1,6,8,4,5]
z=[7,4,6,7,3]

plt.Figure(figsize=(9,4))

plt.subplot(2,2,2)
##plt.subplot(1,2,1)  1 row 2 column , where constructing

plt.plot(x,y, color='blue')
```

```python
##barplot

categories = [ 'a' , 'b' , 'c', 'd']
values = [5,3,5,6]

plt.bar(categories, values , color = 'pink')
```

```python
##histogram
data=[1,2,3,3,4,4,5,5,2,2,3,3,5,5,3,3,2,2]
plt.hist(data, bins=5,edgecolor='black' )
```

```python
##scatterplot

x=[1,2,3,4,5]
y=[1,6,8,4,5]


plt.scatter(x,y)
```

```python
##piechart

##need lables

lables = ['a', 'b' , 'c','d','e']
size=[20,10,50,15,5]
colors = ['red','blue','green','pink','purple']
explode = (.2,0.1,0.4,0.2,0.25)

plt.pie(size,labels=lables,explode=explode)
```
